```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Initialize lists for Name, Year, Version, and Price
Name = []
Year = []
Version = []
Price = []

# Iterate through each result in results
for result in results:
    # Extract Name from the <h2> tag
    name_tag = result.find('h2')
    name_text = name_tag.get_text(strip=True) if name_tag else "N/A"
    Name.append(name_text)

    # Extract Year
    year_tag = result.find('span', {'class': '_18b01e88'})  # Replace 'year_class' with the actual class name
    year_text = year_tag.get_text(strip=True) if year_tag else "N/A"
    Year.append(year_text)

    # Extract Version
    version_tag = result.find('span', {'class': '_948d9e0a _36ddd253'})  # Replace 'version_class' with the actual class name
    version_text = version_tag.get_text(strip=True) if version_tag else "N/A"
    Version.append(version_text)

    # Extract Price
    price_tag = result.find('span', {'class': '_1f2a2b47'})  # Replace 'price_class' with the actual class name
    price_text = price_tag.get_text(strip=True) if price_tag else "N/A"
    Price.append(price_text)

# Create a DataFrame with the extracted data
data = {
    'Name': Name,
    'Year': Year,
    'Version': Version,
    'Price': Price
}
df = pd.DataFrame(data)

# Display the DataFrame
print(df)
"""df = pd.DataFrame(data)

# Save the DataFrame to an Excel file
df.to_excel('Data Mined.xlsx', index=False)

print("Data Mined.xlsx")
"""
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Lists to store the data
Name = []
Year = []
Version = []
Price = []

# Pagination loop for the first 13 pages
for i in range(1, 14):
    # Website URL with pagination
    website = f'your Website={i}'

    # Send an HTTP GET request to the website
    response = requests.get(website)

    # Check if the request was successful
    if response.status_code != 200:
        print(f"Failed to fetch page {i}. Status code: {response.status_code}")
        continue  # Skip to the next page if the current one fails

    # Parse the HTML content with BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find all listings on the page
    results = soup.find_all('article', {'class': '_68441e28'})

    # Extract information from each listing
    for result in results:
        # Extract Name from the <h2> tag
        name_tag = result.find('h2')
        name_text = name_tag.get_text(strip=True) if name_tag else "N/A"
        Name.append(name_text)

        # Extract Year
        year_tag = result.find('span', {'class': '_18b01e88'})  # Adjust class name if necessary
        year_text = year_tag.get_text(strip=True) if year_tag else "N/A"
        Year.append(year_text)

        # Extract Version
        version_tag = result.find('span', {'class': '_948d9e0a _36ddd253'})  # Adjust class name if necessary
        version_text = version_tag.get_text(strip=True) if version_tag else "N/A"
        Version.append(version_text)

        # Extract Price
        price_tag = result.find('span', {'class': '_1f2a2b47'})  # Adjust class name if necessary
        price_text = price_tag.get_text(strip=True) if price_tag else "N/A"
        Price.append(price_text)

    print(f"Page {i} scraped successfully.")

# Create a DataFrame with the extracted data
data = {
    'Name': Name,
    'Year': Year,
    'Version': Version,
    'Price': Price
}
df = pd.DataFrame(data)

# Save the DataFrame to a CSV file
df.to_csv('Data Mined.csv', index=False)

print("Data scraping completed. Results saved to 'Data Mined.csv'.")
```
